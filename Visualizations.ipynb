```python
plt.figure(figsize=(15,9))

plt.bar(totalvacdf.Country, totalvacdf["Cumulative Persons Fully Vaccinated"].sort_values(ascending = False), color="#29335C", align="center")

plt.xticks(rotation=0)

plt.title("Cumulative Persons Fully Vaccinated by Country", fontsize=15, fontweight="bold")
plt.xlabel("Country",  fontsize=16)
plt.ylabel("Cumulative Persons Fully Vaccinated",  fontsize=16)
plt.yscale("log")

plt.grid(axis="y", alpha = 0.45)

plt.savefig("Visuals/Cumulative Persons Fully Vaccinated by Country.png")

plt.show()
```
